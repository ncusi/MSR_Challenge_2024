```python
def print_low_high(latest_sharings_df):
    print_bootstrap_low_high(latest_sharings_df, 'author_number')
    print_bootstrap_low_high(latest_sharings_df, 'commit_number')
    print_bootstrap_low_high(latest_sharings_df, 'forks_count')
    print_bootstrap_low_high(latest_sharings_df, 'stargazers_count')
```

```python
def print_bootstrap_low_high(df, column):
    low, high = bootstrap((df[column],), np.std, confidence_level=0.95).confidence_interval
    print(round(low, 2))
    print(round(high, 2))
```

```python
print_low_high(latest_pr_sharings)
```

```python
print_low_high(latest_commit_sharings)
```
