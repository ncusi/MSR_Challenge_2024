```python
import json
```

```python
with open('repository_statistics_git.json') as repository_statistics_git_file:
    repository_statistics_git = json.load(repository_statistics_git_file)
```

```python
with open('repository_statistics_github.json') as repository_statistics_github_file:
    repository_statistics_github = json.load(repository_statistics_github_file)
```

```python
repository_statistics_git
```

```python
import pandas as pd
```

```python
repository_statistics_git_df = pd.DataFrame.from_dict(repository_statistics_git, orient='index')
```

```python
repository_statistics_github_df =  pd.DataFrame.from_dict(repository_statistics_github, orient='index')
```

```python
repository_statistics_git_df.head()
```

```python
repository_statistics_github_df.head()
```

```python
repository_statistics_git_df.shape
```

```python
repository_statistics_github_df.shape
```

```python
unified_statistics_df = pd.merge(repository_statistics_git_df, repository_statistics_github_df, left_index=True, right_index=True)
```

```python
unified_statistics_df.shape
```

```python
unified_statistics_df.head()
```

```python
unified_statistics_df.loc[unified_statistics_df['in_latest_pr_sharings_y']==True,'Source']='PR'
```

```python
unified_statistics_df.loc[unified_statistics_df['in_latest_commit_sharings_y']==True,'Source']='Commit'
```

```python
unified_statistics_df.loc[(unified_statistics_df['in_latest_pr_sharings_y']==True)&(unified_statistics_df['in_latest_commit_sharings_y']==True),'Source']='PR and Commit'
```

```python
unified_statistics_df['Source'].describe()
```

```python
pr_commit_statistics_df = unified_statistics_df = unified_statistics_df[(unified_statistics_df['Source'].isin(['PR', 'Commit', 'PR and Commit']))]
```

```python
pr_commit_statistics_df.shape
```
