```python
import json
```

```python
from pathlib import Path
```

```python
commit_sharings_path = Path('./DevGPT/snapshot_20231012/20231012_230826_commit_sharings.json')
```

```python
pr_sharings_path = Path('./DevGPT/snapshot_20231012/20231012_233628_pr_sharings.json')
```

```python
issue_sharings_path = Path('./DevGPT/snapshot_20231012/20231012_235128_issue_sharings.json')
```

```python
with open(commit_sharings_path) as commit_sharings_file:
    commit_sharings = json.load(commit_sharings_file)
```

```python
with open(pr_sharings_path) as pr_sharings_file:
    pr_sharings = json.load(pr_sharings_file)
```

```python
with open(issue_sharings_path) as issue_sharings_file:
    issue_sharings = json.load(issue_sharings_file)
```

```python
commit_sharings_sources = commit_sharings['Sources']
pr_sharings_sources = pr_sharings['Sources']
issue_sharings_sources = issue_sharings['Sources']
```

```python
import pandas as pd
```

```python
commit_chat_sharings = []
```

```python
for commit_sharings_source in commit_sharings_sources:
    commit_chat_sharings.extend(commit_sharings_source['ChatgptSharing'])
```

```python
len(commit_chat_sharings)
```

```python
pr_chat_sharings = []
for pr_sharings_source in pr_sharings_sources:
    pr_chat_sharings.extend(pr_sharings_source['ChatgptSharing'])
```

```python
issue_chat_sharings = []
for issue_sharings_source in issue_sharings_sources:
    issue_chat_sharings.extend(issue_sharings_source['ChatgptSharing'])
```

```python
len(pr_chat_sharings)
```

```python
len(issue_chat_sharings)
```

```python
commit_chat_sharings_df = pd.DataFrame.from_records(commit_chat_sharings)
pr_chat_sharings_df = pd.DataFrame.from_records(pr_chat_sharings)
issue_chat_sharings_df = pd.DataFrame.from_records(issue_chat_sharings)
```

```python
all_chat_sharings_df = pd.concat([commit_chat_sharings_df,pr_chat_sharings_df,issue_chat_sharings_df])
```

```python
all_chat_sharings_df.shape
```

```python
commit_chat_sharings_df.shape[0] + pr_chat_sharings_df.shape[0] + issue_chat_sharings_df.shape[0]
```

```python
all_chat_sharings_df = all_chat_sharings_df.drop_duplicates(subset='URL', keep="last")
```

```python
all_chat_sharings_df = all_chat_sharings_df.dropna()
```
